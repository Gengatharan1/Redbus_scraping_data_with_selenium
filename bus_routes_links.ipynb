### ANDHRA PRADESH STATE ROAD TRANSPORT CORPORATION (APSRTC) BUS ROUTES AND LINKS

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# open the chrome browser
driver_AP = webdriver.Chrome()

# get the URL of the webpage
driver_AP.get(state_links[0])

time.sleep(3)

# Maximize window
driver_AP.maximize_window()

# retrive bus links and route
wait = WebDriverWait(driver_AP, 20)     # load the webpage upto 20 sec

def andhra_route_link(path, max_pages): 
    # Initialize lists
    LINKS_AP = []
    ROUTE_AP = []

    # depends on the pagination of the webpage
    for i in range(1, max_pages + 1):
        paths = driver_AP.find_elements(By.XPATH, path)
        
        for element in paths:
            LINKS_AP.append(element.get_attribute("href"))
            ROUTE_AP.append(element.text)
            
        try:
            # Wait for the pagination element to be present
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')
            driver_AP.execute_script("arguments[0].scrollIntoView();", next_button)
            time.sleep(3)
            next_button.click()
            
        except NoSuchElementException:
            print(f"No more pages to paginate at step {i}")
            break
            
    return LINKS_AP, ROUTE_AP

LINKS_AP, ROUTE_AP = andhra_route_link("//a[@class='route']", 5)
```

```python
# converting list into Data frame

df_AP = pd.DataFrame({"Route_name":ROUTE_AP, "Route_link":LINKS_AP})
df_AP
```

### TELANGANA BUS ROUTES AND LINKS

```python
# Open the browser
driver_TG = webdriver.Chrome()

# Get a URL webpage of the Telangana
driver_TG.get(state_links[1])

driver_TG.maximize_window()

time.sleep(3)

wait = WebDriverWait(driver_TG, 20)

def telengana_route_link(path, max_pages):   
    LINKS_TELANGANA = []
    ROUTE_TELANGANA = []
    
    for i in range(1, max_pages+1):
        paths = driver_TG.find_elements(By.XPATH, path)
        
        # retrive the route links 
        for links in paths:
            l = links.get_attribute("href")
            LINKS_TELANGANA.append(l)
            
        # retrive names of the routes
        for route in paths:
            ROUTE_TELANGANA.append(route.text)
            
        try:
            # Wait for the pagination element to be present
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')
            driver_TG.execute_script("arguments[0].scrollIntoView();", next_button)
            time.sleep(3)
            next_button.click()
            
        except NoSuchElementException:
            print(f"No more pages to paginate at step {i}")
            break
            
    return LINKS_TELANGANA, ROUTE_TELANGANA

LINKS_TELANGANA, ROUTE_TELANGANA = telengana_route_link("//a[@class='route']", 3)
```
